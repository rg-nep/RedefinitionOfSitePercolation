## defined fundtions

```python
def get_cum_freq(a, bins=100, normed=True):
    """
        Cumulative frequency where x is linearly spaced
        
        parameters
        ----------
        a      : numpy array like
                 of numbers which (<= 1)
        bins   : int, optional
                 number of points
        normed : boolean, optional
        
        returns
        -------
        a tuple of x and y coordinates
            
    """
    ps = np.linspace(0, 1, bins);
    count = []
    for p in ps:
        b = len(a[a<p])
        count.append(b)
    y = np.array(count)
    if normed:
        y = y / b
    return ps, y
```

### Estimating pc via scaling solution

$(p-p_c) L^{1/\nu} = const$

$ (p_c(L) - p_c(\infty)) L^{1/\nu} = const $

$ p_c(L) - p_c(\infty) = L^{-1/\nu}$

$ p_c(L) = p_c(\infty) + L^{-1/\nu}$

```python
def find_deviation_in_range(one_by_nu, start=-1, stop=0.5, num=5):
    deviations = []
    for at_x in np.linspace(start, stop, num):
        heights = []
        for L in df.columns:
            p, y = df[L]
            x = (p- pc)*L**(one_by_nu)
            a = len(x[x <= at_x])
            heights.append([x[a], y[a]])
            pass

        heights = np.array(heights)
#         print(heights)
        ##now find max and min value of heights
        a = np.max(heights[:,1])
        b = np.min(heights[:,1])
        diff = a-b
#         print("max-min = ?")
#         print("{} - {} = {}".format(a, b, diff))
        deviations.append([at_x, diff])
#         deviations.append([at_x, np.std(heights[:,1])])
        pass
    return deviations
```

```python
print("format : [exponent, x, deviation]")
deviations_for_different_exponent = []
for ex in np.linspace(0.65, 0.8, 100):
    devs = find_deviation_in_range(ex, x1[0], x1[1], 100)
    deviations_for_different_exponent.append([ex, devs])
    pass
# deviations_for_different_exponent
```

```python
## finding minimum standard deviation for exponent
stds = []
for a in deviations_for_different_exponent:
#     print(a)
    b = np.array(a[1])
#     print(b)
    stds.append([a[0], np.std(b[:,1])])
    pass
stds = np.array(stds)
# print(stds)
m = np.min(stds[:,1])
print(m)
print("minimum median for exponent")
print(stds[stds[:,1] <= m])
```
